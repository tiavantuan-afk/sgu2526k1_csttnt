## Nội dung bài
Hiển thị ngẫu nhiên $N=50$ hình tròn:
+ Tâm $c_i = (x_i,y_i)$ ngẫu nhiên với $x_i, y_i \in \left[0, 1 \right]$
+ Màu sắc $color_i$ ngẫu nhiên với $color_i \in \left[0, 1 \right]$
+ Diện tích $area_i = \pi * r_i^2$ với sinh ngẫu nhiên $r_i \in \left[0, 1 \right]$


## Bài làm

```python
N = 50 # Số lượng hình tròn

# Sinh tọa độ tâm (x, y) trong khoảng [0, 1]
x = np.random.rand(N)
y = np.random.rand(N)

# Sinh màu sắc ngẫu nhiên (giá trị từ 0 đến 1)
colors = np.random.rand(N)

# Sinh diện tích ngẫu nhiên
# Bán kính r từ 0 đến 15 (nhân 15 để hình to rõ hơn trên biểu đồ)
r = 15 * np.random.rand(N)
area = np.pi * (r ** 2)

# 3. Kiểm tra dữ liệu 
print(f'Đã sinh {N} hình tròn:')
print(f'- Tọa độ X (min-max): {min(x):.2f} - {max(x):.2f}')
print(f'- Tọa độ Y (min-max): {min(y):.2f} - {max(y):.2f}')
print(f'- Diện tích lớn nhất: {max(area):.2f}')

# 4. Hiển thị biểu đồ (Scatter Plot)
plt.figure(figsize=(8, 6)) # Thiết lập kích thước khung hình
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='viridis')

# Trang trí biểu đồ
plt.title('Biểu đồ 50 hình tròn ngẫu nhiên')
plt.xlabel('Trục X')
plt.ylabel('Trục Y')
plt.grid(True, linestyle='--', alpha=0.3) # Thêm lưới mờ
plt.colorbar(label='Thang màu') # Hiện thanh màu bên cạnh

# Hiển thị
plt.show()
```

### Sinh ngẫu nhiên hình tròn

```python
%time
# generate randomly the center (x,y ) of circle, color, area
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2  # 0 to 15 point radii
```
